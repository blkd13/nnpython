```python
nTest = int(len(datI)/10)
dInSet  = datI[nTest:] # 入力データ
dOutSet = datO[nTest:] # 出力データ
tInSet  = datI[:nTest] # テスト用入力データ
tOutSet = datO[:nTest] # テスト用出力データ
nIn  = len(datI[0]) # 入力データ次元数
nOut = len(datO[0]) # 出力データ次元数
mult = np.array(dInSet).flatten().mean()
# 出力データの正規化（0〜1）。出力層はtanh、つまり-1〜1しか取れないので。
# minOut = np.min(dOutSet)
# rangeOut = np.max(dOutSet) - minOut
# dOutSetNorm = (dOutSet - minOut ) / rangeOut * 2 - 1
# tOutSetNorm = (tOutSet - minOut ) / rangeOut * 2 - 1

# 2層ある中間層の次元数
nMid=[nIn,200,nOut]
rSeed=5904

# zero初期化（こっちの方が全然速い）
w0 = tf.Variable(tf.truncated_normal([nMid[0],nMid[1]]))
w1 = tf.Variable(tf.zeros([nMid[1],nMid[2]],tf.float32))

b0 = tf.Variable(tf.zeros([nMid[1]],tf.float32))
b1 = tf.Variable(tf.zeros([nMid[2]],tf.float32))

dIn  = tf.placeholder(tf.float32,[None,nIn ])
dOut = tf.placeholder(tf.float32,[None,nOut])

p = tf.nn.sigmoid(tf.matmul(tf.nn.tanh(tf.matmul(dIn,w0)+b0*mult),w1)+b1*mult)
loss = -tf.reduce_sum(dOut*tf.log(p)+(1-dOut)*tf.log(1-p))
# loss = tf.reduce_sum(tf.square(dOut-tf.nn.tanh(tf.matmul(tf.nn.tanh(tf.matmul(dIn,w0)+b0),w1)+b1)))
train_step = tf.train.AdamOptimizer().minimize(loss)
# train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
correct_prediction = tf.equal(tf.sign(p-0.5),tf.sign(dOut-0.5))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

sess = tf.Session()
sess.run(tf.global_variables_initializer())
loss_val = []
test_val = []
i = 0
```

```python
nCycle = 100
nDelta = 100
print(datetime.now().strftime("%Y/%m/%d %H:%M:%S") + " Start")
for _ in range(nCycle):
    i += 1
    sess.run(train_step, feed_dict={dIn:dInSet, dOut:dOutSet})
    if i % nDelta == 0:
        loss_val.append([i/nDelta,sess.run(loss, feed_dict={dIn:dInSet, dOut:dOutSet})])
        test_val.append([i/nDelta,sess.run(loss, feed_dict={dIn:tInSet, dOut:tOutSet})])
        print(datetime.now().strftime("%Y/%m/%d %H:%M:%S") + '  Step: %d, Loss: %f' % (i, loss_val[len(loss_val)-1][1]))
print(datetime.now().strftime("%Y/%m/%d %H:%M:%S") + " End ")
```
